```python
# Python and Pandas Assessment Notebook

import pandas as pd
import numpy as np
from functools import reduce

# Task 1: Create a DataFrame using Pandas
# Create a DataFrame with the following data:
# Name: Alice, Bob, Charlie, David
# Age: 25, 30, 35, 28
# City: New York, San Francisco, Los Angeles, Chicago

# Your code here:
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 28],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)

print("Task 1: DataFrame")
print(df)


# Task 2: Row and Column Manipulation
# Drop the 'City' column from the DataFrame created in Task 1

# Your code here:
df_no_city = df.drop('City', axis=1)

print("\nTask 2: DataFrame without City")
print(df_no_city)

# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them

# Your code here:
df_null = pd.DataFrame({
    'Name': ['Alice', None, 'Charlie'],
    'Age': [25, None, 35],
    'City': [None, 'Toronto', 'Ottawa']
})

print("\nTask 3: Original DataFrame with Nulls")
print(df_null)

df_filled = df_null.fillna({
    'Name': 'Unknown',
    'Age': 0,
    'City': 'Not Specified'
})

print("\nAfter Filling Null Values")
print(df_filled)

# Task 4: GroupBy and Describe
# Using the following DataFrame, group by 'Category' and describe the 'Value' column

df_group = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Value': [10, 20, 15, 25, 30, 35]
})

# Your code here:
group_stats = df_group.groupby('Category')['Value'].describe()

print("\nTask 4: GroupBy and Describe")
print(group_stats)

# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Your code here:
vertical_concat = pd.concat([df1, df2], ignore_index=True)

print("\nTask 5: Vertical Concatenation")
print(vertical_concat)

horizontal_merge = pd.concat([vertical_concat, df3.reindex(vertical_concat.index)], axis=1)

print("\nHorizontal Merge")
print(horizontal_merge)

# Task 6: Tuples and Sets
# Create a tuple of fruits and a set of numbers.
# Then, try to add an element to each and observe the difference.

# Your code here:

fruits = ('Apple', 'Banana', 'Orange')
numbers_set = {1, 2, 3, 4, 5}

print("\nTask 6")
print("Tuple:", fruits)
print("Set:", numbers_set)

# Sets can be modified
numbers_set.add(6)
print("Updated Set:", numbers_set)

print("Tuples are immutable and cannot be modified directly.")
# Task 7: Dictionaries
# Create a dictionary of student names and their scores.
# Then, update a student's score and add a new student.

# Your code here:
students = {
    'John': 85,
    'Jane': 90,
    'Bob': 78
}

students['Jane'] = 95
students['Alice'] = 88

print("\nTask 7: Dictionary")
print(students)

# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.
# Then, use a lambda function to do the same.

# Your code here:

def square(num):
    return num ** 2

square_lambda = lambda x: x ** 2

print("\nTask 8")
print(square(4))
print(square(7))

print(square_lambda(4))
print(square_lambda(7))
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
class EvenIterator:
    def __init__(self):
        self.current = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < 5:
            value = self.current
            self.current += 2
            self.count += 1
            return value
        raise StopIteration

print("\nTask 9: Iterator")

for num in EvenIterator():
    print(num)

def even_generator():
    for i in range(0, 10, 2):
        yield i

print("Generator Output")

for num in even_generator():
    print(num)

# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:
squared = list(map(lambda x: x**2, numbers))

product = reduce(lambda x, y: x * y, numbers)

evens = list(filter(lambda x: x % 2 == 0, numbers))

print("\nTask 10")
print("Squared:", squared)
print("Product:", product)
print("Even Numbers:", evens)

# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect1 = Rectangle(5, 3)
rect2 = Rectangle(8, 4)

print("\nTask 11")
print("Rectangle 1 Area:", rect1.area())
print("Rectangle 1 Perimeter:", rect1.perimeter())

print("Rectangle 2 Area:", rect2.area())
print("Rectangle 2 Perimeter:", rect2.perimeter())
# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:

avg_salary = df_employees.groupby('Department')['Salary'].mean()

print("\nTask 12: Average Salary by Department")
print(avg_salary)

high_salary = df_employees[df_employees['Salary'] > 60000]['Name']

print("\nEmployees with Salary > 60000")
print(high_salary)

df_employees['Bonus'] = df_employees['Salary'] * 0.10

print("\nEmployee Data with Bonus")
print(df_employees)
```
